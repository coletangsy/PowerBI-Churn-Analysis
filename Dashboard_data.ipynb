```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv("./data/telecom-churn_dummies.csv")
model = pickle.load(open('telecom_churn_best_model.sav', 'rb'))
```

```python
df.head()
```

```python
df.drop(columns=["Churn","Unnamed: 0"],inplace=True)
df.head()
```

```python
df2 = df[["Contract_Month-to-month","TotalCharges","OnlineSecurity_No","TechSupport_No","OnlineBackup_No","PaymentMethod_Electronic check", "DeviceProtection_No","Contract_Two year","tenure_group_1 - 12","MonthlyCharges"]]
```

```python
df2
```

```python
df["Predicted_Churn"] = model.predict(df2)
```

```python
df["Predicted_Churn_rate"] = model.predict_proba(df2)[:,1]
```

```python
df.to_csv("telecom-churn_dashboard.csv",index=False)
```
